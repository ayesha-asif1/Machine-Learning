```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create DataFrame
data = {
    "OrderID": [101, 102, 103, 104, 105, 106],
    "Product": ["Headphones", "T-Shirt", "Laptop Stand", "Notebook", "Mouse", "BackPack"],
    "Category": ["Electronics", "Clothing", "Accessories", "Stationery", "Electronics", "Accessories"],
    "Price": [25, 45, 78, 12, 78, 90],
    "Quantity": [1, 2, 3, 4, 5, 7],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Chicago", "Miami"]
}

df = pd.DataFrame(data)
df["Total"] = df["Price"] * df["Quantity"]

# Discount & Final Price
df["Discount"] = df["Total"] * 0.15
df["FinalPrice"] = df["Total"] - df["Discount"]

print(df)
```

```python
product_sales=df.groupby("Product")["Total"].sum()
plt.bar(df["Product"],df["Total"])
plt.title("Total sales by product")
plt.xlabel("Product")
plt.ylabel("Total Revenue($)")
plt.show()
```

<h3>Pie Chart</h3>

```python
plt.pie(df["Total"],
        labels=df["City"])
plt.title("Sales by city(Simple pie chart)")
plt.show()
```

```python
# Advance pie chart

city_sales=df.groupby("City")["Total"].sum()

plt.pie(
    city_sales,
    labels=city_sales.index,
    autopct="%1.1f%%",
    startangle=90,
    shadow=True,
    explode=[0.1]*len(city_sales)
)
plt.title("Sales share by city")
plt.show()
```
